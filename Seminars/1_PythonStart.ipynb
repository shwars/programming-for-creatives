```python
import jturtle as jt

for j in range(36):
  for i in range(4):
    jt.forward(100)
    jt.right(90)
  jt.right(10)
jt.done()
```

```python
def draw(x,n):
  if n==0:
    jt.forward(x)
  else:
    draw(x/3,n-1)
    jt.left(60)
    draw(x/3,n-1)
    jt.left(60+180)
    draw(x/3,n-1)
    jt.left(60)
    draw(x/3,n-1)

jt.right(90)
for i in range(3):
  draw(100,4)
  jt.right(120)
jt.done()
```

## Анализ данных

```python
import pandas as pd

df = pd.read_excel('titanic.xlsx')
df
```

```python
df['survived'].mean()
```

```python
df.groupby('sex')['survived'].mean()
```

```python
df.groupby('sex')['survived'].mean().plot(kind='bar')
```

```python
df.groupby('pclass')['survived'].mean().plot(kind='bar')
```

```python
import matplotlib.pyplot as plt
fig,ax = plt.subplots(1,2,figsize=(10,3))
for i,x in enumerate(['male','female']):
  ax[i].set_title(x)
  df[df['sex']==x].groupby('pclass')['survived'].mean().plot(kind='bar',ax=ax[i])
```
